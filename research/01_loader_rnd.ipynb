## PyPDFLoader Evaluation

### Overview

PyPDFLoader was tested on the SBI Health Insurance Retail policy document containing 27 pages.

### Performance

* Total Pages Extracted: 27
* Processing Time: 1.39 seconds

### Text Extraction Quality

* Text was extracted successfully from all pages.

* Content was readable without major extraction failures.

### Metadata Extraction

Available metadata:

* Author
* Title
* Source Path
* Total Pages
* Creation Date
* Page Number

### Advantages

* Easy to use.
* Good text extraction quality.
* Suitable for standard insurance policy documents.
* Integrates directly with LangChain pipelines.

### Limitations

* Slower compared to PyMuPDFLoader.
* Repeated company headers appear in extracted text.
* Formatting noise exists in some sections.
* Table structure is not preserved perfectly.

### Conclusion

PyPDFLoader provides reliable extraction for insurance documents and can be used as a baseline loader. However, performance optimization may be required for large document collections.

```python
import pdfplumber
import time

PDF_PATH = "../data/general/health/SBI_Health_Insurance_Retail.pdf"

start = time.time()

plumber_pages = []

with pdfplumber.open(PDF_PATH) as pdf:
    for i, page in enumerate(pdf.pages):
        text = page.extract_text()
        tables = page.extract_tables()
        plumber_pages.append({
            "page": i,
            "text": text,
            "tables": tables
        })

end = time.time()

print("Loader: PDFPlumber")
print("Pages:", len(plumber_pages))
print("Time:", round(end - start, 2), "seconds")
print("Tables on first page:", len(plumber_pages[0]["tables"]))
print("\nSample Text:\n")
print(repr(plumber_pages[0]["text"][:2000] if plumber_pages[0]["text"] else ""))
```

## PDFPlumber Evaluation

### Overview
PDFPlumber was tested on the SBI Health Insurance Retail policy document containing 27 pages.

### Performance
- Total Pages Extracted: 27
- Processing Time: 4.01 seconds
- Tables Detected on First Page: 0

### Text Extraction Quality
- Text was extracted, but the first line showed repeated characters.
- Example issue: `SSSSBBBBIIII GGGGeeeennnneeeerrrraaaallll`
- Main policy text was readable after the noisy header.
- Bullet points were extracted, but special bullet symbols appeared as unicode characters.

### Table Extraction
- No tables were detected on the first page.
- For this specific PDF, PDFPlumber did not provide a table extraction advantage.

### Advantages
- Can be useful for PDFs with clear table structures.
- Allows page-level extraction and table extraction separately.
- Useful for deeper table-specific experiments.

### Limitations
- Slowest among tested loaders so far.
- Header text extraction was noisy.
- No table detected on the first page.
- Extra cleaning would be required before using this output in RAG.

### Conclusion
PDFPlumber is not preferred as the default loader for this SBI Health Insurance PDF because it was slower and introduced character duplication noise. It may still be useful as a fallback for PDFs where table extraction is important.
